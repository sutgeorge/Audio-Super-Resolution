```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

training_set_values, validation_set_values, testing_set_values = np.array([]), np.array([]), np.array([])
training_files, validation_files, testing_files = DatasetGenerator.split_list_of_files()
high_res_training_files = training_files[1]

training_minimum_value, training_maximum_value = None, None

for training_file in high_res_training_files:
    print(training_file)
    chunk = np.load("preprocessed_dataset/high_res/" + training_file)
    chunk = chunk.reshape(chunk.shape[0])
    training_minimum_value = np.min(chunk) if training_minimum_value == None else min(training_minimum_value, np.min(chunk))
    training_maximum_value = np.max(chunk) if training_maximum_value == None else max(training_maximum_value, np.max(chunk))
    print(chunk.shape)

print("Minimum value in training set: {}".format(training_minimum_value))
print("Maximum value in training set: {}".format(training_maximum_value))
```
